```python
df=data[['DateTime_x','EquipmentClassification','Description_Equipment','Description_MeasPt','Description_AnalParm','DataValue','RPM']]
```

```python
df.sort_values(by = ['DateTime_x','Description_MeasPt'], inplace=True)
df[1:15]
```

```python
DateTime=df['DateTime_x'].tolist()
EquipmentClassification=df['EquipmentClassification'].tolist()
Description_Equipment=df['Description_Equipment'].tolist()
Description_AnalParm=df['Description_AnalParm'].tolist()
Description_MeasPt=df['Description_MeasPt'].tolist()
DataValue=df['DataValue'].tolist()
```

```python
tab=[[DateTime[0], EquipmentClassification[0], Description_Equipment[0],Description_MeasPt[0],DataValue[0]
      ,DataValue[1],DataValue[2],DataValue[3],DataValue[4],DataValue[5],DataValue[6],np.nan,np.nan,np.nan,np.nan,RPM[0]]]


for i in range (1,len(DateTime)-11):
    row=[np.nan]*16
    
    if DateTime[i-1]!=DateTime[i]: #or Description_MeasPt[i-1]!=Description_MeasPt[i]:
        row[0]=DateTime[i]
        row[1]=EquipmentClassification[i]
        row[2]=Description_Equipment[i]
        row[3]=Description_MeasPt[i]
        row[15]=RPM[i]
        
        for n in range (11):
            if Description_Equipment[i]==Description_Equipment[i+n]: #and Description_MeasPt[i]==Description_MeasPt[i+n]:
                if Description_AnalParm[i+n]=='TEMPO P-P     ':
                    row[4]=DataValue[i+n]
                elif Description_AnalParm[i+n]=='1xRPM         ':
                    row[5]=DataValue[i+n]
                elif Description_AnalParm[i+n]=='2-10xRPM      ':
                    row[6]=DataValue[i+n]
                elif Description_AnalParm[i+n]=='10-60xRPM     ':
                    row[7]=DataValue[i+n]
                elif Description_AnalParm[i+n]=='ISO 10-1000HZ ':
                    row[8]=DataValue[i+n]
                elif Description_AnalParm[i+n]=='TEMPO PIC     ':
                    row[9]=DataValue[i+n]
                elif Description_AnalParm[i+n]=='HFD           ':
                    row[10]=DataValue[i+n]
                elif Description_AnalParm[i+n]=='TEMPOREL P-P  ':
                    row[11]=DataValue[i+n]
                elif Description_AnalParm[i+n]=='10-80xRPM     ':
                    row[12]=DataValue[i+n]
                elif Description_AnalParm[i+n]=='10-100xRPM    ':
                    row[13]=DataValue[i+n]
                elif Description_AnalParm[i+n]=='ISO 2-1000HZ  ':
                    row[14]=DataValue[i+n]
        tab.append(row)

data_x=pd.DataFrame(tab, columns = ['DateTime', 'EquipmentClassification','Description_Equipment','Description_MeasPt',
                                    'TEMPO P-P', '1xRPM', '2-10xRPM',
                                    '10-60xRPM', 'ISO 10-1000HZ', 'TEMPO PIC',
                                   'HFD', 'TEMPOREL P-P', '10-80xRPM',
                                   '10-100xRPM', 'ISO 2-1000HZ','RPM'])
data_x
```

## Join Reports

```python
reports=pd.read_csv("C:/Users/aurel/OneDrive/Documents/ENSIAME-INSA/Cours/S8/Projet4A/DataForLAMIH/ClientA_Reports.csv", index_col=0)
```

```python
reports.sort_values(by = 'SurveyDate',inplace=True)
```

```python
conditions = [(reports['FaultStatus'] =='1. FOR INFORMATION'), (reports['FaultStatus'] =='2. NO INTERVENTION REQUIRED'),
              (reports['FaultStatus'] =='3. ATTENTIVE FOLLOW UP'),(reports['FaultStatus'] =='4. PROACTIVE INTERVENTION')
             ,(reports['FaultStatus'] =='5. CORRECTIVE INTERVENTION')]
choices = [1, 2, 3,4,5]
reports['FaultStatus'] = np.select(conditions, choices)
reports.sort_values(by = 'SurveyDate',inplace=True)
reports
```

```python
def addlabel(dfe,rapport):
    
    df=dfe
    df['SurveyDate']=["None" for i in range (df.shape[0])]
    df['FaultStatus']=[0 for i in range (df.shape[0])]
    df['FaultName']=["None" for i in range (df.shape[0])]
    df['Comment']=["None" for i in range (df.shape[0])]
    df['Location']=["None" for i in range (df.shape[0])]
    date=rapport['SurveyDate'].tolist()
    equipmentid=rapport['EquipmentID'].tolist()
    faultstatus=rapport['FaultStatus'].tolist()
    faultname=rapport['FaultName'].tolist()
    comment=rapport['Comment'].tolist()
    location=rapport['Location'].tolist()
    
    
    for k in range (rapport.shape[0]):
        
        for i in range (df.shape[0]):
        
            if (date[k][:9] in df.iat[i,0]) and (equipmentid[k] in df.iat[i,2]):
            
                df.iat[i,17]=date[k]
                df.iat[i,16]=faultstatus[k]
                df.iat[i,18]=faultname[k]
                df.iat[i,19]=comment[k]
                df.iat[i,20]=location[k]
            
    
    return df
```

```python
data=addlabel(data_x,reports)
data
```
